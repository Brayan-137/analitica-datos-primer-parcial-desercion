# 📺 Deserción de estudiantes

## Contexto y Variables Principales

Los datos utilizados corresponden a un estudio sobre deserción y éxito académico en educación superior. Cada registro representa a un estudiante de una institución de educación superior, integrando información obtenida de varias bases de datos institucionales. El conjunto incluye 3,630 estudiantes y 36 variables predictoras, con información disponible desde el momento de la matrícula y también sobre el desempeño académico en los primeros semestres.

Se dispone de variables relacionadas con el perfil académico previo 🎓 (tipo de formación anterior, nota de calificación previa, orden y modalidad de aplicación, programa al que ingresa), características demográficas y personales 🌍 (nacionalidad, estado civil, asistencia diurna/nocturna), así como factores familiares y socioeconómicos 👨‍👩‍👧‍👦💼 (nivel educativo y ocupación de la madre y del padre, condición de deudor, beca, pago de matrícula al día). También se incluyen indicadores del desempeño universitario temprano 📚 (créditos cursados, aprobados, evaluaciones y calificaciones en el primer y segundo semestre), que permiten capturar señales iniciales del progreso académico del estudiante.

El objetivo principal es predecir la variable de respuesta Target, que clasifica a los estudiantes en tres categorías: Dropout y Graduate. Este problema se plantea como una tarea de clasificación, con desbalance entre categorías, y resulta especialmente relevante porque permite identificar de manera temprana a estudiantes en riesgo de deserción ⚠️, apoyar la toma de decisiones institucionales 🏫 y diseñar estrategias de acompañamiento más efectivas para mejorar la permanencia y el éxito académico.

## Punto 2:

- Ahora debes abordar el tratamiento de valores faltantes y la detección de datos atípicos en el dataset. Explica brevemente la importancia de estos pasos en el análisis de decersión, describe las técnicas que emplearás para imputar o eliminar valores perdidos, y justifica el método elegido para identificar y manejar outliers. Documenta cada decisión y muestra ejemplos concretos aplicados sobre el dataframe.

```python
import pandas as pd

def tratar_outliers_iqr(df, columnas):
    resumen = []
    df_tratado = df.copy()
    
    for col in columnas:
        Q1 = df_tratado[col].quantile(0.25)
        Q3 = df_tratado[col].quantile(0.75)
        IQR = Q3 - Q1
        
        limite_inf = Q1 - 1.5 * IQR
        limite_sup = Q3 + 1.5 * IQR
        
        outliers = df_tratado[(df_tratado[col] < limite_inf) | (df_tratado[col] > limite_sup)]
        n_outliers = len(outliers)
        
        # Reemplazar outliers por los límites (Winsorizing)
        df_tratado[col] = df_tratado[col].clip(lower=limite_inf, upper=limite_sup)
        
        resumen.append({
            'Variable'       : col,
            'Q1'             : round(Q1, 4),
            'Q3'             : round(Q3, 4),
            'IQR'            : round(IQR, 4),
            'Límite inferior': round(limite_inf, 4),
            'Límite superior': round(limite_sup, 4),
            'Outliers found' : n_outliers,
            '% sobre total'  : round(n_outliers / len(df) * 100, 2)
        })
    
    resumen_df = pd.DataFrame(resumen)
    
    print("=" * 60)
    print("RESUMEN DE TRATAMIENTO DE OUTLIERS — IQR")
    print("=" * 60)
    print(f"Tamaño original del DataFrame : {df.shape[0]} filas x {df.shape[1]} columnas")
    print(f"Tamaño tratado del DataFrame  : {df_tratado.shape[0]} filas x {df_tratado.shape[1]} columnas")
    print(f"Variables tratadas            : {len(columnas)}")
    print(f"Total outliers reemplazados   : {resumen_df['Outliers found'].sum()}")
    print("=" * 60)
    print("\nDetalle por variable:\n")
    print(resumen_df.to_string(index=False))
    
    return df_tratado, resumen_df

# Uso
df_tratado, resumen = tratar_outliers_iqr(df, [
    'Previous qualification (grade)',
    'Admission grade',
    'Unemployment rate',
    'Inflation rate',
    'GDP'
])
```

### Tratamiento de Valores Atípicos (Outliers)
Para la identificación de outliers, se empleó el Rango Intercuartílico (IQR), un método no paramétrico que permite la detección de anomalías sin necesidad de asumir una distribución normal en los datos. En cuanto al tratamiento, se optó por la técnica de Winsorización, reemplazando los valores atípicos por los límites correspondientes del intervalo. Esta decisión se tomó con el fin de minimizar la pérdida de información que supondría la eliminación de filas completas.

Cabe destacar que este procedimiento se aplicó a todas las variables continuas, con excepción de las unidades curriculares del primer y segundo semestre. Se decidió conservar los valores originales en estas variables debido a que su variabilidad y valores extremos se consideran factores determinantes y de alta relevancia para explicar el fenómeno de la deserción.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Selección de variables (continuas + target)

variables_seleccionadas = [
    'Curricular units 2nd sem (approved)',
    'Curricular units 1st sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Curricular units 1st sem (grade)',
    'Tuition fees up to date',
    'Scholarship holder',
    'Debtor',
    'Age at enrollment',
    'Admission grade',
    'Previous qualification (grade)'
]

X_train_sel = X_train[variables_seleccionadas]
X_test_sel  = X_test[variables_seleccionadas]

# Analisis de multicolinealidad

scaler_vif = StandardScaler()
X_scaled_vif = scaler_vif.fit_transform(X_train_sel)

vif_df = pd.DataFrame({
    'Variable': variables_seleccionadas,
    'VIF': [variance_inflation_factor(X_scaled_vif, i)
            for i in range(X_scaled_vif.shape[1])]
}).sort_values('VIF', ascending=False)

print("=" * 45)
print("ANÁLISIS DE MULTICOLINEALIDAD (VIF)")
print("=" * 45)
print(vif_df.to_string(index=False))
print("\n⚠️  VIF > 10 indica multicolinealidad severa.")
```

En esta sección deberás evaluar el desempeño del modelo de regresión logística construido. Para ello, divide el conjunto de datos en bases de entrenamiento y test, ajusta el modelo sobre la base de entrenamiento y evalúa su capacidad predictiva sobre la base de test. Presenta métricas como exactitud, sensibilidad, especificidad y curva ROC, interpretando los resultados obtenidos.

```python
from sklearn.metrics import (
    accuracy_score, recall_score, f1_score,
    roc_auc_score, roc_curve,
    confusion_matrix, ConfusionMatrixDisplay,
    classification_report
)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

CLASES = ['Dropout', 'Graduate']  # 0 → Dropout, 1 → Graduate


# PREDICCIONES

y_pred       = pipeline.predict(X_test_sel)
y_pred_proba = pipeline.predict_proba(X_test_sel)[:, 1]


# MÉTRICAS GLOBALES

accuracy     = accuracy_score(y_test, y_pred)
sensibilidad = recall_score(y_test, y_pred, pos_label=1)
especificidad = recall_score(y_test, y_pred, pos_label=0)
f1           = f1_score(y_test, y_pred, average='weighted')
auc          = roc_auc_score(y_test, y_pred_proba)

print("=" * 45)
print("MÉTRICAS DE DESEMPEÑO DEL MODELO")
print("=" * 45)
print(f"  Exactitud (Accuracy)           : {accuracy:.4f}")
print(f"  Sensibilidad (Recall Graduate) : {sensibilidad:.4f}")
print(f"  Especificidad (Recall Dropout) : {especificidad:.4f}")
print(f"  F1-Score (ponderado)           : {f1:.4f}")
print(f"  AUC-ROC                        : {auc:.4f}")
print("=" * 45)
print()
print(classification_report(y_test, y_pred, target_names=CLASES))


#  VISUALIZACIONES

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# A) Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=CLASES)
disp.plot(ax=axes[0], cmap='Blues', colorbar=False)
axes[0].set_title('Matriz de Confusión', fontweight='bold', fontsize=12)

# B) Curva ROC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
axes[1].plot(fpr, tpr, color='steelblue', lw=2, label=f'AUC = {auc:.4f}')
axes[1].plot([0, 1], [0, 1], color='gray', linestyle='--', lw=1, label='Clasificador aleatorio')
axes[1].fill_between(fpr, tpr, alpha=0.1, color='steelblue')
axes[1].set_xlabel('Tasa de Falsos Positivos (1 - Especificidad)')
axes[1].set_ylabel('Tasa de Verdaderos Positivos (Sensibilidad)')
axes[1].set_title('Curva ROC', fontweight='bold', fontsize=12)
axes[1].legend(loc='lower right')
axes[1].set_xlim([0, 1])
axes[1].set_ylim([0, 1.02])

# C) Distribución de probabilidades por clase real
axes[2].hist(y_pred_proba[y_test == 0], bins=30, alpha=0.6, color='tomato',
             label='Dropout real', edgecolor='white')
axes[2].hist(y_pred_proba[y_test == 1], bins=30, alpha=0.6, color='steelblue',
             label='Graduate real', edgecolor='white')
axes[2].axvline(0.5, color='black', linestyle='--', linewidth=1, label='Umbral = 0.5')
axes[2].set_xlabel('Probabilidad predicha (Graduate)')
axes[2].set_ylabel('Frecuencia')
axes[2].set_title('Distribución de Probabilidades por Clase', fontweight='bold', fontsize=12)
axes[2].legend()

plt.suptitle('Evaluación del Modelo de Regresión Logística', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()


# VALIDACIÓN CRUZADA

from sklearn.model_selection import cross_validate, StratifiedKFold

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

resultados_cv = cross_validate(
    pipeline,
    X_train_sel, y_train,
    cv=cv,
    scoring=['accuracy', 'f1_weighted', 'roc_auc'],
    return_train_score=True
)

cv_df = pd.DataFrame({
    'Fold'          : range(1, 6),
    'Accuracy (val)': resultados_cv['test_accuracy'].round(4),
    'F1 (val)'      : resultados_cv['test_f1_weighted'].round(4),
    'AUC (val)'     : resultados_cv['test_roc_auc'].round(4),
})

print("=" * 55)
print("VALIDACIÓN CRUZADA — 5 FOLDS (StratifiedKFold)")
print("=" * 55)
print(cv_df.to_string(index=False))
print("-" * 55)
print(f"  Accuracy promedio : {resultados_cv['test_accuracy'].mean():.4f} ± {resultados_cv['test_accuracy'].std():.4f}")
print(f"  F1 promedio       : {resultados_cv['test_f1_weighted'].mean():.4f} ± {resultados_cv['test_f1_weighted'].std():.4f}")
print(f"  AUC promedio      : {resultados_cv['test_roc_auc'].mean():.4f} ± {resultados_cv['test_roc_auc'].std():.4f}")
print("=" * 55)
```
